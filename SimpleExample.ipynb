Let's write an equation:

$$ f(x) = \sin x $$

Now some simple graphs

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi)
plt.plot(x, np.sin(x))
plt.xlabel('$x$', fontsize=14)
plt.ylabel('$\sin(x)$', fontsize=14)
plt.grid()
```
